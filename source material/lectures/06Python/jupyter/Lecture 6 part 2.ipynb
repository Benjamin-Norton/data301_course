# Loops

## `while` loops
A `while` loop repeats statements while a condition is True.

```python
n=1
while n <= 5:
    print(n) 
    n=n+1
```

## Shorthand

```python
x = 1
x = x + 5
print(x)

# equivalent to 
x = 1
x += 5
print(x)
```

```python
n = 1
while n <= 5:
    print(n)
    n += 1 
```

## `for` loops
A `for` loop repeats statements a given number of times.

```python
# notice how 6 is not included in this sequence
for i in range(1,6):
    print(i)
```

```python
# increment by 2 starting from 1 and ending at 9
for i in range(1, 10, 2):
    print(i)
```

```python
# increment by 2 starting from 2 and ending at 9
for i in range(2, 10, 2):
    print(i)
```

```python
# prints the numbers 5 to 1
for i in range(5,0, -1):
    print(i)
```

```python
# notice how the default start is 0
for i in range(4):
    print(i)
```

The `for` loop is like a short-hand for the `while` loop:

```python
i=0
while i < 10:
    print(i)
    i += 1
```

```python
for i in range(0, 10, 1):
     print(i)
```

**Infinite loops** are caused by an incorrect loop condition or not updating values within the loop so that the loop condition will eventually be false.

```python
n = 1
while n <= 5:
    print(n)
```

A common mistake is to be off by one.  How can we fix the following code to print 0 to 10?

```python
for i in range(0,10):
    print(i)
```

```python
# solution
for i in range(0,11):
    print(i)
```

### Example 30
Write a program that asks the user for 5 positive numbers and prints the maximum, sum, and average of the numbers.

```python
total = 0
maxnum = float("-inf")
m = 5
for i in range(1,m+1):
    n = input("Enter a number: ")
    n = float(n)
    total += n # total = total + n
    if n > maxnum:
        maxnum = n

print("Max:",maxnum)
print("Sum:",total)
print("Avg:",total/m)
```

Alternatively, we could save all these values to a list and use the `max` and `sum` functions

```python
userlist = []
for i in range(1,6):
    n = input("Enter a number: ")
    userlist.append(float(n))
```

```python
print(userlist)
print("Max:", max(userlist))
print("Sum:", sum(userlist))
print("Avg:", sum(userlist)/len(userlist))
```

# Functions and Procedures
A **procedure** is a sequence of program statements that have a specific task that they perform.
A **function** is a procedure that returns a value after it is executed.

```python
def doubleNum(x):
    x = x*2
    print("Number :",x)
    return x
doubleNum(5)
```

```python
def addf(x, y):
    return x + y 

print(addf(2,5))
print(addf("2","5"))
```

```python
def plus2(x):
    x + 2
print(plus2(3))
nothing = plus2(3)
print(nothing)
```

```python
# returning multiple values in a function using a list
def plus2(x):
    out = x + 2
    return [x,out]

print(plus2(3))
print(type(plus2(3)))
```

```python
# returning multiple values in a function using a tuple
def plus2(x):
    out = x + 2
    return (x,out)

returned_values = plus2(3)
print(returned_values)
print(type(returned_values))
print(returned_values[0], "+ 2 =", returned_values[1])
```

```python
# notice how we use multiple return statemnets here
def gradeLetter(pgrade):
    if (pgrade >= 80):
        return "A"
    elif (pgrade >= 68):
        return "B"
    elif (pgrade >= 55):
        return "C"
    elif (pgrade >= 50):
        return "D"
    else: 
        return "F"

print('80=',gradeLetter(80))
print('68=',gradeLetter(68))
print('55=',gradeLetter(55))
print('50=',gradeLetter(50))
print('43=',gradeLetter(43))
```

```python
# returns grade in percent and letter
def grade(pgrade):
    if (pgrade >= 80):
        grade = "A"
    elif (pgrade >= 68):
        grade = "B"
    elif (pgrade >= 55):
        grade = "C"
    elif (pgrade >= 50):
        grade = "D"
    else: 
        grade = "F"
    return [grade, pgrade]

print(grade(81))
print(grade(70))
print(grade(60))
print(grade(54))
print(grade(30))
```

## Lambda functions

 Lambda functions are functions that do not have a name.  There are useful for creating one-time use functions

```python
def doFunc(func, val):
    return func(val)

print(doFunc(doubleNum, 10))
print(doFunc(lambda x: x * 3, 5))
```
